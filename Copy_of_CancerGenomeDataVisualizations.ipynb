```python
#10:25pm
import pandas as pd
import plotly.express as px
from dash import jupyter_dash
from jupyter_dash import JupyterDash
from dash import dcc, html
from dash.dependencies import Input, Output

mutations_df = pd.DataFrame(filled_df)

# Initialize the Dash app
app = JupyterDash(__name__)

app.layout = html.Div([
    html.H1("Gene Mutations Stacked Bar Chart"),

    dcc.Graph(id='stacked-bar-chart')
])

@app.callback(
    Output('stacked-bar-chart', 'figure'),
    Input('stacked-bar-chart', 'id')
)
def update_stacked_bar_chart(_):
    # Count the number of mutations for each gene and mutation type
    mutation_counts = mutations_df.groupby(['Hugo_Symbol', 'One_Consequence']).size().reset_index(name='Mutation_Count')

    # Calculate the total number of mutations per gene
    total_mutations = mutation_counts.groupby('Hugo_Symbol')['Mutation_Count'].sum().reset_index(name='Total_Mutations')

    # Merge the total mutations back to the mutation counts
    mutation_counts = mutation_counts.merge(total_mutations, on='Hugo_Symbol')

    # Sort by the total number of mutations
    mutation_counts = mutation_counts.sort_values(by='Total_Mutations', ascending=False)

    # Set the order of genes based on the sorted total mutations
    gene_order = mutation_counts['Hugo_Symbol'].unique()

    # Create a stacked bar chart
    fig = px.bar(
        mutation_counts,
        x='Hugo_Symbol',
        y='Mutation_Count',
        color='One_Consequence',
        title='Number of Mutations per Gene colored by Mutation Type',
        labels={'Mutation_Count': 'Number of Mutations'},
        category_orders={'Hugo_Symbol': gene_order}
    )
    return fig
# Run the app
app.run()

```

```python
import pandas as pd
import plotly.express as px
from dash import jupyter_dash
from jupyter_dash import JupyterDash
from dash import dcc, html
from dash.dependencies import Input, Output

# Load your datasets (assuming they are already preprocessed and cleaned)
df = merged_df
mutations_df = filled_df

# Initialize the Dash app
app = JupyterDash(__name__)

app.layout = html.Div([
    html.H1("Cancer Genome Patient Data Dashboard"),

    html.Div([
        html.Label("Select Variable for Distribution:"),
        dcc.Dropdown(
            id='distribution-var',
            options=[{'label': col, 'value': col} for col in ['gender', 'race_list', 'ethnicity', 'tumor_tissue_site']],
            value='gender'
        )
    ]),

    dcc.Graph(id='distribution-plot'),

    html.Div([
        html.Label("Select X and Y variables for Scatter Plot:"),
        dcc.Dropdown(
            id='scatter-x-var',
            options=[{'label': col, 'value': col} for col in df.columns],
            value='days_to_last_followup'
        ),
        dcc.Dropdown(
            id='scatter-y-var',
            options=[{'label': col, 'value': col} for col in df.columns],
            value='age_at_initial_pathologic_diagnosis'
        )
    ]),

    dcc.Graph(id='scatter-plot'),

    html.Div([
        html.Label("Select Mutation Gene for Analysis:"),
        dcc.Dropdown(
            id='gene-var',
            options=[{'label': gene, 'value': gene} for gene in df['Hugo_Symbol'].unique()],
            value='KDM1A'
        )
    ]),

    dcc.Graph(id='mutation-plot'),

    html.Div([
        html.Label("Select Patient ID for Mutation Analysis:"),
        dcc.Dropdown(
            id='patient-id-var',
            options=[{'label': pid, 'value': pid} for pid in mutations_df['Patient_ID'].unique()],
            value=mutations_df['Patient_ID'].unique()[0]
        )
    ]),

    dcc.Graph(id='stacked-bar-chart')
])

@app.callback(
    Output('distribution-plot', 'figure'),
    Input('distribution-var', 'value')
)
def update_distribution_plot(selected_var):
    # Ensure the selected variable exists in the dataframe
    if selected_var not in df.columns:
        return {}
    fig = px.histogram(df, x=selected_var, title=f'Distribution of {selected_var}')
    return fig

@app.callback(
    Output('scatter-plot', 'figure'),
    [Input('scatter-x-var', 'value'),
     Input('scatter-y-var', 'value')]
)
def update_scatter_plot(x_var, y_var):
    # Ensure the selected variables exist in the dataframe
    if x_var not in df.columns or y_var not in df.columns:
        return {}
    fig = px.scatter(df, x=x_var, y=y_var, title=f'{y_var} vs {x_var}')
    return fig

@app.callback(
    Output('mutation-plot', 'figure'),
    Input('gene-var', 'value')
)
def update_mutation_plot(gene):
    gene_df = df[df['Hugo_Symbol'] == gene]
    if gene_df.empty:
        return {}
    fig = px.bar(gene_df, x='Variant_Classification', y='Tumor_Sample_Barcode', title=f'Variant Classification for {gene}')
    return fig

@app.callback(
    Output('stacked-bar-chart', 'figure'),
    Input('patient-id-var', 'value')
)
def update_stacked_bar_chart(patient_id):
    # Filter the mutation data for the selected patient
    patient_df = mutations_df[mutations_df['Patient_ID'] == patient_id]

    # Group by gene and consequence to get the count of mutations
    mutation_counts = patient_df.groupby(['Hugo_Symbol', 'One_Consequence']).size().reset_index(name='count')

    # Create a stacked bar chart
    fig = px.bar(
        mutation_counts,
        x='Hugo_Symbol',
        y='count',
        color='One_Consequence',
        title=f'Number of Mutations per Gene for Patient {patient_id}',
        labels={'count': 'Number of Mutations'}
    )
    return fig

# Run the app
app.run_server(mode='external')
```
